```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error
```

```python
#Small artificial dataset
X=np.array([
      [1,2,10],
      [2,1,12],
      [3,3,9],
      [4,5,11],
      [5,4,13],
      [6,6,8],
      [7,7,7],
      [8,8,14]
])

#y mainly depends on XX1 & X2
y=np.array([6,7,12,18,20,24,28,32])
```

```python
df= pd.DataFrame(X,columns=["X1","X2","X3"])
df["y"]=y
print("Dataset : \n",df)
```

```python
#----------------------------------------
# 1. Without Regularization
#----------------------------------------
lr = LinearRegression()
lr.fit(X, y)
y_pred_lr = lr.predict(X)

print("\nLinear Regression (No Regularization)")
print("coefficients:", lr.coef_)
print("MSE:", mean_squared_error(y, y_pred_lr))
```

```python
print("Intercept:", lr.intercept_)

```

```python
#----------------------------------------
# 2. Ridge Regression (L2)
#----------------------------------------

ridge = Ridge(alpha=1.0)   # alpha = regularization strength
ridge.fit(X, y)
y_pred_ridge = ridge.predict(X)

print("\nRidge Regression (L2 Regularization)")
print("Coefficients:", ridge.coef_)
print("MSE:", mean_squared_error(y, y_pred_ridge))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

lasso = Pipeline([
    ("scaler", StandardScaler()),
    ("lasso", Lasso(alpha=0.5))
])

lasso.fit(X, y)

print("\nLasso Regression (L1 Regularization)")
print("Coefficients:", lasso.named_steps["lasso"].coef_)
print("MSE:", mean_squared_error(y, lasso.predict(X)))

```
