```python
import numpy as np
```

```python
X1=np.array([1,2])
###X2=np.array([1,2,2])
y=np.array([1,3])
```

```python
X=np.column_stack((np.ones(2),X1))
```

```python
B=np.linalg.inv(X.T@ X) @ X.T@y
```

```python
b0,b1=B
```

```python
print("intercept (b0)",b0)
print("coefficient for studied hour (b1)",b1)
##print("Coefficient for sleep hours(b2)",b2)
```

```python
#prediction
y_pred=b0+b1*X1
print("predicted y : ",y_pred)
```

```python
for lam in [0, 1, 5, 100]:
    I = np.eye(X.shape[1])
    I[0, 0] = 0   # don't regularize the intercept

    B = np.linalg.inv(X.T @ X + lam * I*I ) @ X.T @ y
    print(f"Lambda = {lam}, coefficients = {B}")
```
