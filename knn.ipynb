```python
import math
from collections import Counter
```

```python
#setup data set
point =['a','b','c','d']
x1=[7,8,5,9]
x2=[2,1,4,6]
labels=['pos','neg','pos','neg']
p=[8,2]
k=3

#calculate euclidian distance
dataset=[]
for i in range(len(point)):
    dist = math.sqrt(((p[0]-x1[i])**2)+((p[1]-x2[i])**2))
    print(point[i],dist)
    dataset.append({'name': point[i],'dist': dist,'lebel': labels[i]})
```

```python
#sort by distance (closest first)
dataset.sort(key=lambda x: x['dist'])
```

```python
neighbors = dataset[:k]

print(f"The {k} nearest neighbors to point {p} are:")
print("-" * 30)
for n in neighbors:
    print(f"neighbor {n['name']}: distance {n['dist']:.2f}, class: {n['lebel']}")
print("-" * 30)
```

```python
print(f"the {k} nearest neighbour to point {p} are :")
for neighbor in neighbors:
  print(f"point {neighbor['name']} with distance {neighbor['dist']:.2f}")
```
